# **iLOC**

```python
import pandas as pd

df = pd.read_csv("data.csv")
```

```python

df.head()
```

```python
df.iloc[0]  # row_number
```

```python
df.iloc[0,1]  # row_number with column number
```

```python
df.iloc[2:4]  # start_row: end_row
```

```python
df.iloc[2:4,0:2]  # start_row: end_row , start_col: end_col
```
