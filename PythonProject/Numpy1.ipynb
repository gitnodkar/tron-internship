```python
import numpy as np

#append

a=np.array([1,2,3,4])
print(a)

x=np.append(a,99) # add element to array
print(x)

b=np.array([[1,2],[3,4]])
print(b)

c=np.append(b,100)
print(c)
```

```python
#insert()

c=np.array([1,2,3,4,5])
print(c)

d=np.insert(c,2,10) # insert value into array in single position
print(d)

e=np.insert(c,(0,3),12) # insert value into array in multiple positions
print(e)
```

```python
#delete()

d=np.delete(c,2) # delete value from array. ---> new_var=np.delete(var_name, position)
print(d)
```

```python
#where()

x=np.array([11,12,13,14,15,16,12,18])
print(x)

y=np.where(x==12) # find the position of data
print(y)

y=np.where(x==200)
print(y)
```

```python
#array_split()
x=np.array([11,12,13,14,15,16,12,18])

y=np.array_split(x,3)
print(y)

y=np.array_split(x,12)
print(y)
```

```python
#asarray(var_name , dtype= datatype)

y=np.asarray(x,dtype=float) # used to change datatype of array
print(y)

y=np.asarray(x,dtype=complex)
print(y)
```

```python
#unique()

t=np.array([1,2,3,4,5,6,2,3,7,5,8,90,8])
print(t)

y=np.unique(t) # Unique elements
print(y)

y=np.unique(t,return_index=True) # unique elements with position
print(y)

y=np.unique(t,return_index=True,return_counts=True) # unique elements with position and count
print(y)
```

```python
# functions in array
# 1. zeros() only zeros decimal
# 2. ones() only ones decimal
# 3. full()
# 4. eye() diagonal matrix (identity)
```

```python
#zeros()

x=np.zeros(3)
print(x)

x=np.zeros((2,3))
print(x)
```

```python
#ones()

x=np.ones(5)
print(x)

x=np.ones((3,3))
print(x)
```

```python
#eye()

x=np.eye(3) # Output is always 2D matrix
print(x)
print(x.ndim)
```

```python
#eye()

x=np.full((3,3),10)
print(x)

x=np.full((2,3),'a')
print(x)
```

```python
#copy()
import numpy as np

x=np.array([11,14,15,99,100])

y=x.copy() # Both variable id address will be different
print(x)

#indexing --->
print(x[1])   #14

#var_name[position]=value
x[1]=1111
print(x)
print(y)

# print(id(x))

# print(y)
# print(id(y))
```

```python
#view()

x=np.array([11,14,15,99,100])

y=x.view() # Modification in one also affect the another
x[1]=111
print(x)
print(y)

y[3]=2000
print(y)
print(x)
```

```python
#split()
x=np.array([1,2,3,4,5,6,7,8])
print(x)

y=np.array_split(x,4)
print(y)

y1=np.split(x,4) # equal separation of the array is needed
print(y1)


#y1=np.split(x,3) # Error cannot divide in equal elements in array
#print(y1)
```

```python
#add two arrays
#concatenate()
x=np.array([11,56,778,99])
y=np.array([22,55,77,44])

z=np.concatenate((x,y))
print(z)
```

```python
#stack()

z1=np.stack((x,y)) # Convert in 2D dimensional array
print(z1)

z1=np.stack((x,y),axis=0) # Arrange as Row
print(z1)

z1=np.stack((x,y),axis=1) # Arrange as Columns
print(z1)
```

```python
#hstack()

z2=np.hstack((x,y)) # Convert into rows format
print(z2)
```

```python
#vstack()

z2=np.vstack((x,y)) # Convert into column format
print(z2)
```

```python
#dstack()

z2=np.dstack((x,y)) # Convert into Height Format
print(z2)

#Convert to 3D dimensional
print(z2.ndim)
```

```python
#Arithmetic Operatipns in Numpy arrays

# a+b ----> add()

x=np.array([1,2,3])
y=np.array([1,2,3])

z=np.add(x,y)
print(z)
```

```python
# a-b ----> subtract()

z=np.subtract(x,y)
print(z)
```

```python
# a*b ----> multiply()

z=np.multiply(x,y)
print(z)
```

```python
# a/b ----> divide()

z=np.divide(x,y)
print(z)
```

```python
# a%b ----> mod()

z=np.mod(x,y)
print(z)
```

```python
# a**2 ----> power()

z=np.power(x,y)
print(z)
```

```python
# Scalar Operations

x=np.array([10,20,30,40])
print(x)

#previous_var=previous_var+number
#previous_var+=number

#x=x+5
x+=5
print(x)

#x=x-5
x-=5
print(x)

#x=x/5
x=x/5
print(x)

```
